# Matplotlib Interface for Rivet (June 28)

The user calls the `rivet-mkhtml` command to create a web page with all the plots (same as before). The user specifies the yoda file(s), the overall style of the plots as a string (rivet is the default), and an optional yaml file that contains the modified rcParams (which overwrites the style):

```python
rivet-mkhtml mc1.yoda mc2.yoda --style rivet --rcParams rcParams.yaml
```

`rivet-mkhtml` then calls `rivet-makeyaml` which creates a yaml file for each plot that contains:
- The histogram data (Key: histograms)
- The individual plot features given from the .plot files (Key: plot_features)
- The overall style of the plots (Key: plot_style)
- The modified rcParams (Key: rcParams)

We now have a yaml file for each plot that contains the histogram data, individual plot features, the plot style, and any modified rcparams. To create the plots,  `rivet-plot` is called for each yaml file generated by `rivet-makeyaml`. E.g.,:

```python
rivet-plot ATLAS_d10-x01-y01.yaml
```

Finally, once all the plots are outputted, the html web page is created.

# Other Use Cases
Instead of calling `rivet-mkhtml`, the user could just create a single plot using two commands. The first one, 

```python
rivet-makeyaml mc1.yoda mc2.yoda --style rivet --rcParams rcParams.yaml
```

outputs the yaml file for each plot. Then, the user can plot an individual file (in this case ATLAS_d10-x01-y01.yaml):

```python
rivet-plot ATLAS_d10-x01-y01.yaml
```
